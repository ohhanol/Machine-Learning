##CSV 파일 읽어오기
- Pandas 라이브러리로 Dataframe 생성해서 csv 파일로 저장하기
  - Dataframe
  
  - csv 파일


````python
# pandas 라이브러리 설치 (불러오기)
!pip install pandas
````

````python
# 라이브러리 import (읽어오기)
import pandas as pd
import numpy as np
import os
````

````python
# 데이터프레임 생성 (입력된 데이터, 열, 행에 대한 데이터프레임)
data = [[4, 5, 6],
        [7, 8, 9],
        [10, 11, 12]]

columns = ['a', 'b', 'c']
index = [1, 2, 3]

df = pd.DataFrame(data, index = index ,columns = columns)
````

````python
# df 확인
print(df)
````

````python
# 특정 행, 열 읽어오기
print(df.iloc[:,0]) # 첫번째 열 읽어오기
print(df.iloc[0]) # 첫번째 행 읽어오기
````

````python
# dataframe 복사 및 수정
df2 = df.copy() # dataframe의 복사본 생성
df2["a"] = [11, 22, 33] # dataframe의 a 열 값 수정
````

````python
print(df)
print(df2)
````

````python
# df 저장
df.to_csv(os.path.join(colab_path, 'df.csv'))
````
